```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random
import math
import csv

#definindo o número máximo de display do output para 150.
#essa mudança é apenas estética, não alterando os valores que serão guardados,
#apenas permitindo ver todas as linhas, sem "pular" as intermediárias.
pd.set_option('display.max_rows',150)

#fazendo a leitura do arquivo .csv (tabela), usando a biblioteca pandas
iris = pd.read_csv("../input/iris/Iris.csv")
#vendo as 5 primeiras linhas do dataframe
iris.head()
#vendo todas as informações do dataframe
#iris
#vendo quantas amostras existem de cada espécie
#iris["Species"].value_counts()
```

```python
#lendo o dataset Iris.csv
with open("../input/iris/Iris.csv") as orig:
    #data recebe cada linha de f (dataset Iris.csv)
    data = [line.strip() for line in orig]
#removendo o primeiro elemento da matriz, pois estes eram os nomes das colunas do dataframe
#(Id, SepalLengthCm, Species, etc)
# e com esses elementos, não seria possível transformar os valores do dataframe em floats
data.pop(0)
dados = []
for line in data:
    dados.append( [(x) for x in line.split(',')])
for i in range(0,len(dados)):
    for j in range (0,len(dados[i])-1):
        dados[i][j] = float(dados[i][j])
dados.pop()
#printando a matriz de dados resultante (todos os elementos, exceto pelos nomes das espécies, agora são floats)
dados
```

```python
#Agora, iremos separar os dados para treino
t = len(dados)
tamanho = int(0.7*t) #para sabermos o tamanho da amostra, que é 70% dos dados
#é um número inteiro pois não temos como avaliar metade de um dado
amostra = random.choices(dados,k=tamanho)
amostra.sort(key=lambda x: x[4]) #organizando os dados por ordem crescente da PetalWidth, p/ facilitar a visualização
amostra
```

```python
#Separando os dados para teste (30%, excluindo os 70% selecionados para treino)
teste = []
for i in dados:
    if i not in amostra: #=só vai pegar os dados que não foram selecionados para treino
        teste.append(i)
teste
```

Para o algoritmo de KNN, fazemos a análise da distância euclidiana ou de Manhattan. Para qualquer um dos métodos escolhidos, o procedimento é análise as variáveis necessárias e ver quais são os elementos com maior proximidade do selecionado.

Os elementos relevantes para a nossa análise serão SepalLengthCm, SepalWidthCm, PetalLengthCm, e PetalWidthCm. Para facilitar a escrita do código, essas variáveis serão chamadas de sl,sw,pl, e pw, respectivamente.

```python
def d_euc(amostra, distancia, sl, sw, pl, pw):
    #percorrendo todos os elementos de amostra e calculando suas respectivas distâncias
    #amostra[i][n], 1<=n<=4 -> elementos SepalLengthCm, SepalWidthCm, PetalLengthCm, e PetalWidthCm
    #temos que comparar todos os elementos!
    #o único elemento "ignorado" é o Id (amostra[i][0])
    for i in range(0,len(amostra)):
        dist = math.sqrt( (amostra[i][1]-sl)**2 + (amostra[i][2]-sw)**2 + (amostra[i][3]-pl)**2 + (amostra[i][4]-pw)**2)
        distancia.append(dist)
```

```python
def d_mtt(amostra,distancia,sl,sw,pl,pw):
    #percorrendo todos os elementos de amostra e calculando suas respectivas distâncias
    #amostra[i][n], 0<=n<=4 -> elementos SepalLengthCm, SepalWidthCm, PetalLengthCm, e PetalWidthCm
    #temos que comparar todos os elementos!
    #o único elemento "ignorado" é o Id (amostra[i][0])
    for i in range(0,len(amostra)):
        dist = abs(amostra[i][1]-sl) + abs(amostra[i][2]-sw) + abs(amostra[i][3]-pl) +abs(amostra[i][4]-pw)
        distancia.append(dist)
```

```python
print("Deseja fazer a análise de forma manual (inserção de valures de um espécime) ou automática (análise de todos os espécimes de teste)?")
op2 = int(input("0=Manual, 1=Automática. Escolha ="))

if op2 ==0:
    print("Por favor insira os valores de SepalLengthCm, SepalWidthCm, Petal LengthCm, e PetalWidthCm.")
    sl = float(input("SepalLengthCm = "))
    sw = float(input("SepalWidthCm = "))
    pl = float(input("PetalLengthCm = "))
    pw = float(input("PetalPetalWidthCm = "))
op = int(input("Selecione o método Euclidiano ou de Manhattan para a análise da distância.\n2=Euclidiano, 3=Manhattan.\nEscolha="))
distancia = []

erro = 0
if op2==0:
    if op == 2:
        print("Calculando a distância Euclidiana...")
        d_euc(amostra,distancia,sl,sw,pl,pw)
        menor = min(distancia)
        menor_dist = distancia.index(menor)
        print("O elemento com menor distância tem uma distância de ",distancia[menor_dist])
        print("Esse elemento é uma ",amostra[menor_dist][5])
    else:
        print("Calculando a distância de Manhattan...")
        d_mtt(amostra,distancia,sl,sw,pl,pw)
        menor = min(distancia)
        menor_dist = distancia.index(menor)
        print("O elemento com menor distância tem uma distância de ",distancia[menor_dist])
        print("Esse elemento é uma ",amostra[menor_dist][5])
else:
    if op == 2:
        print("Calculando a distância Euclidiana...")
        for i in range(0,len(teste)):
            print("Elemento: ",teste[i])
            d_euc(amostra,distancia,teste[i][1],teste[i][2],teste[i][3],teste[i][4])
            menor = min(distancia)
            menor_dist = distancia.index(menor)
            print("Menor distância:",menor,"\nElemento com menor distancia: ",menor_dist)
            print("O elemento com menor distância tem uma distância de",distancia[menor_dist])
            print("Esse elemento é uma",amostra[menor_dist][5])
            if(amostra[menor_dist][5]!=teste[i][5]):
                erro = erro + 1
            print("---------")
            distancia = []
        print("Número de erros:",erro,"\nTaxa de acertos:",100-(erro*100/len(teste)),"%")
    else:
        print("Calculando a distância Manhattan...")
        for i in range(0,len(teste)):
            print("Elemento: ",teste[i])
            d_mtt(amostra,distancia,teste[i][1],teste[i][2],teste[i][3],teste[i][4])
            menor = min(distancia)
            menor_dist = distancia.index(menor)
            print("Menor distância:",menor,"\nElemento com menor distancia: ",menor_dist)
            print("O elemento com menor distância tem uma distância de",distancia[menor_dist])
            print("Esse elemento é uma",amostra[menor_dist][5])
            if(amostra[menor_dist][5]!=teste[i][5]):
                erro = erro + 1
            print("---------")
            distancia = []
        print("Número de erros:",erro,"\nTaxa de acertos:",100-(erro*100/len(teste)),"%")
        
```
